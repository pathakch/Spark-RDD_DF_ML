```python
from pyspark import SparkConf
from pyspark import SparkContext
```

```python
sc=SparkContext.getOrCreate(SparkConf().setMaster("local[*]"))
```

```python
sc
```

```python
rdd1=sc.parallelize(['chandan','pathak','sonu','monu','chandan','pathak','munna','munna'])
```

```python
rdd1.collect()
```

```python
rdd1.take(4)
```
